# 07 Rate Scaling


Compare physical-qubit estimates under different encoding-rate assumptions. This is intentionally simple: it is a first-pass scaffold, not a full reproduction.

```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

FIG_DIR = Path('../figures')
FIG_DIR.mkdir(exist_ok=True)
```

```python
logical_qubits = np.arange(16, 1025, 16)
rates = {
    'surface-like low rate': 0.02,
    'intermediate rate': 0.08,
    'high-rate code': 0.20,
}

rows = []
for label, rate in rates.items():
    for k in logical_qubits:
        rows.append({'logical_qubits': k, 'encoding_model': label, 'rate': rate, 'physical_qubits': np.ceil(k / rate)})

df = pd.DataFrame(rows)
df.head()
```

```python
fig, ax = plt.subplots(figsize=(7, 4.5))
for label, g in df.groupby('encoding_model'):
    ax.plot(g['logical_qubits'], g['physical_qubits'], label=label)
ax.set_xlabel('logical qubits')
ax.set_ylabel('physical qubits')
ax.set_title('Encoding rate specifies physical-qubit scale')
ax.legend()
fig.tight_layout()
fig.savefig(FIG_DIR / '07_rate_scaling.png', dpi=180)
plt.show()
```
