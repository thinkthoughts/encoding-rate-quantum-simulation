# 23 Resource Geometry


Use the abstract-level paper estimates as anchor points for lab-report figures. These are not independent reproductions; they are source-derived anchors for communicating the constraint geometry.

```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

FIG_DIR = Path('../figures')
FIG_DIR.mkdir(exist_ok=True)
```

```python
anchors = pd.DataFrame([
    {'simulation': '8x8 TFIM', 'physical_qubits': 2240, 'seconds_per_shot': 200, 'reported_space_reduction_vs_surface_STAR': 5.5},
    {'simulation': 'Fermi-Hubbard', 'physical_qubits': 6300, 'seconds_per_shot': 200, 'reported_space_reduction_vs_surface_STAR': np.nan},
])
anchors
```

```python
fig, ax = plt.subplots(figsize=(6.5, 4.5))
ax.scatter(anchors['physical_qubits'], anchors['seconds_per_shot'], s=120)
for _, row in anchors.iterrows():
    ax.annotate(row['simulation'], (row['physical_qubits'], row['seconds_per_shot']), xytext=(6, 6), textcoords='offset points')
ax.set_xlabel('physical qubits')
ax.set_ylabel('seconds per shot')
ax.set_title('Source-derived resource anchor points')
fig.tight_layout()
fig.savefig(FIG_DIR / '23_resource_geometry.png', dpi=180)
plt.show()
```
