# 13 STAR Parallelism


Model the qualitative benefit of disjoint logical representatives: STAR injections performed in parallel across k logical qubits reduce amortized preparation overhead.

```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

FIG_DIR = Path('../figures')
FIG_DIR.mkdir(exist_ok=True)
```

```python
k = np.arange(1, 257)
serial_cost = k
parallel_cost = np.ones_like(k)
amortized = serial_cost / np.maximum(parallel_cost, 1)

star = pd.DataFrame({'logical_qubits_per_block': k, 'serial_units': serial_cost, 'parallel_units': parallel_cost, 'amortization_factor': amortized})
star.tail()
```

```python
fig, ax = plt.subplots(figsize=(7, 4.5))
ax.plot(star['logical_qubits_per_block'], star['amortization_factor'])
ax.set_xlabel('logical qubits per block, k')
ax.set_ylabel('idealized amortization factor')
ax.set_title('Parallel STAR injection changes the cost geometry')
fig.tight_layout()
fig.savefig(FIG_DIR / '13_star_parallelism.png', dpi=180)
plt.show()
```
